```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import os
os.getcwd()
os.chdir('d:/datasets/d1/Actuarial-Pricing-and-Reserving-GLM/')
```

```python
sdf = pd.read_csv('./data/processed/severity_data.csv')
```

```python
sdf.columns
```

```python
# Fit a Gamma GLM for claim severity
sdf_model = smf.glm(
    formula="claim_amount ~ vehicle_age + region + channel + sum_insured",
    data=sdf,
    family=sm.families.Gamma(sm.families.links.Log())
).fit()

print(sdf_model.summary())
```

```python
# Predict claim severity
sdf['predicted_severity'] = sdf_model.predict()
```
